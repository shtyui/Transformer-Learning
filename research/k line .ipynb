```python
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('VIX0808.csv')

# 确保 'Date' 列是 pandas 的 datetime 格式
df['Date'] = pd.to_datetime(df['Date'])

# 将 'Date' 设置为数据框的索引
df.set_index('Date', inplace=True)

# 自定义时间范围
start_date = '2010-01-04'
end_date = '2010-04-20'
df_subset = df.loc[start_date:end_date]

# 自定义图表大小
figratio = (16, 9)  # 宽高比 (宽度, 高度)
figscale = 1.5  # 放大比例
# 绘制 K 线图并返回 fig 和 ax 对象
fig, axlist = mpf.plot(df_subset, 
                       type='candle', 
                       style='charles', 
                       title='K Line Chart', 
                       ylabel='Price',
                       figratio=figratio, 
                       figscale=figscale,
                       returnfig=True)

# axlist 是一个包含轴对象的列表，提取第一个主轴对象
ax = axlist[0]

custom_title = "Custom K Line Chart Title"
fig.suptitle(custom_title, fontsize=16, fontweight='bold', y=0.95)
# 移动 y 轴到左边
ax.yaxis.tick_left()
ax.yaxis.set_label_position("left")

# 显示图表
plt.show()
```

```python
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('spx0821.csv')

# 确保 'Date' 列是 pandas 的 datetime 格式
df['Date'] = pd.to_datetime(df['Date'])

# 将 'Date' 设置为数据框的索引
df.set_index('Date', inplace=True)

# 自定义时间范围
start_date = '2018-01-01'
end_date = '2018-04-30'
df_subset = df.loc[start_date:end_date]

# 假设VIX数据已经包含在CSV文件中，且列名为'VIX'
# 如果VIX数据在另一个文件或数据框中，需先合并数据
vix_data = df_subset['VIX']  # 提取VIX数据

# 创建 VIX 折线的附加图层
ap = mpf.make_addplot(vix_data, panel=0, color='blue', ylabel='VIX')

# 自定义图表大小
figratio = (16, 9)  # 宽高比 (宽度, 高度)
figscale = 1.5  # 放大比例

# 绘制 K 线图并添加 VIX 折线
fig, axlist = mpf.plot(df_subset, 
                       type='candle', 
                       style='charles', 
                       title='K Line Chart with VIX',
                       ylabel='Price', 
                       figratio=figratio, 
                       figscale=figscale,
                       addplot=ap,  # 添加 VIX 折线
                       returnfig=True)

# axlist 是一个包含轴对象的列表，提取第一个主轴对象
ax = axlist[0]

# 自定义标题
custom_title = "Custom K Line Chart with VIX"
fig.suptitle(custom_title, fontsize=16, fontweight='bold', y=0.95)

# 移动 y 轴到左边
ax.yaxis.tick_left()
ax.yaxis.set_label_position("left")

# 显示图表
plt.show()
```

```python
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('spx0821.csv')

# 确保 'Date' 列是 pandas 的 datetime 格式
df['Date'] = pd.to_datetime(df['Date'])

# 将 'Date' 设置为数据框的索引
df.set_index('Date', inplace=True)

# 自定义时间范围
start_date = '2018-01-01'
end_date = '2018-04-30'
df_subset = df.loc[start_date:end_date]

# 提取VIX和VIX预测值数据
vix_data = df_subset['VIX']  # VIX实际值
vix_predicted_data = df_subset['VIX_Predicted']  # VIX预测值

# 创建 VIX 实际值和预测值的附加图层
ap_vix = mpf.make_addplot(vix_data, panel=0, color='blue', ylabel='VIX')
ap_vix_predicted = mpf.make_addplot(vix_predicted_data, panel=0, color='red', linestyle='--', ylabel='VIX')

# 自定义图表大小
figratio = (16, 9)  # 宽高比 (宽度, 高度)
figscale = 1.5  # 放大比例

# 绘制 K 线图并添加 VIX 和 VIX 预测值折线
fig, axlist = mpf.plot(df_subset, 
                       type='candle', 
                       style='charles', 
                       title='K Line Chart with VIX and Predicted VIX',
                       ylabel='Price', 
                       figratio=figratio, 
                       figscale=figscale,
                       addplot=[ap_vix, ap_vix_predicted],  # 添加 VIX 和 VIX 预测值折线
                       returnfig=True)

# axlist 是一个包含轴对象的列表，提取第一个主轴对象
ax = axlist[0]

# 自定义标题
custom_title = "Custom K Line Chart with VIX and Predicted VIX"
fig.suptitle(custom_title, fontsize=16, fontweight='bold', y=0.95)

# 移动 y 轴到左边
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")

# 显示图表
plt.show()
```
